```python
import requests

city = "London"

r1 = requests.get("https://api.openweathermap.org/data/2.5/forecast",params={"q": city, "units": "metric", "appid": API_KEY})
r1.raise_for_status()
forecast = r1.json()["list"]  
city_info = r1.json()["city"]
lat = city_info["coord"]["lat"]
lon = city_info["coord"]["lon"]

pollution_data = {}
for f in forecast:
    dt = f["dt"]
    r2 = requests.get("https://api.openweathermap.org/data/2.5/air_pollution",params={"lat": lat, "lon": lon, "appid": API_KEY})
    r2.raise_for_status()
    pollution_data[dt] = r2.json()["list"][0]
```

```python
import pandas as pd

pollution_df = pd.DataFrame.from_dict(pollution_data, orient='index')
if 'dt' in pollution_df.columns:
    pollution_df = pd.DataFrame.from_dict(pollution_data, orient='index')
else:
    pollution_df.index.name = 'dt'
    pollution_df = pollution_df.reset_index()
pollution_df
```
